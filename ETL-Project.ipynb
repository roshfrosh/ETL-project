```python
import pandas as pd
import numpy as np
import csv
```

```python
austin_csv = "Resources/Austin_inspections.csv"
austin_df = pd.read_csv(austin_csv)
austin_df.head()
```

```python
austin_df.columns
```

```python
austin_df = austin_df.drop([ "Address" ,"Unnamed: 9", "Inspection Date", "Facility City", "Zip Code","Facility ID", "Process Description"], axis = 1)
austin_df.head()
```

```python
austin_df.index.names=["ID"]
austin_df = austin_df.rename(columns={'Restaurant Name': 'rest_name', 'Common Name': 'rest_no', 'Score':'score'})
Austin_data= austin_df[austin_df["rest_no"].isin(["STARBUCKS", "MCDONALD'S", "BURGER KING", "SUBWAY", "TACO BELL"])]
Austin_data.head()
```

```python
la_csv = "Resources/LA_inspections.csv"
la_df = pd.read_csv(la_csv)
la_df.head()
```

```python
la_df.columns
```

```python
la_df = la_df.drop(['serial_number','grade','facility_city','facility_id','facility_zip', "activity_date", "service_description","service_code", "employee_id", "facility_address", "facility_state", "owner_id", "owner_name", "pe_description", "program_element_pe", "program_name", "program_status", "record_id"], axis = 1)
la_df.head()
```

```python
la_df.index.names=["ID"]
la_df = la_df.rename(columns={'facility_name': 'rest_name', 'Common Name': 'rest_no'})
LA_data= la_df[la_df["rest_no"].isin(["STARBUCKS", "MCDONALD'S", "BURGER KING", "SUBWAY", "TACO BELL"])]
LA_data.head()
```

```python
nyc_csv = 'Resources/NYC_Inspection_Results.csv'
nyc_df = pd.read_csv(nyc_csv)
nyc_df.head
nyc_df.columns
```

```python
nyc_df = nyc_df.drop(['CAMIS', 'BORO', 'BUILDING', 'STREET', 'ZIPCODE',\
       'PHONE', 'CUISINE DESCRIPTION', 'INSPECTION DATE', 'ACTION',\
       'VIOLATION CODE', 'VIOLATION DESCRIPTION', 'CRITICAL FLAG',\
       'GRADE', 'GRADE DATE', 'RECORD DATE', 'INSPECTION TYPE', 'Latitude',\
       'Longitude', 'Community Board', 'Council District', 'Census Tract',\
       'BIN', 'BBL', 'NTA'], axis=1)
nyc_df.head()
```

```python
nyc_df.rename(columns={'DBA':'rest_name', 'Common Name': 'rest_no', 'SCORE':'SCORE'}, inplace=True)
nyc_df.head()
```

```python
# rest_list = nyc_df.loc[nyc_df["DBA"] == "PIZZA HUT" | nyc_df["DBA"] == "STARBUCKS"]
# rest_list
nyc_df = nyc_df[nyc_df['rest_no'].isin(["PIZZA HUT", "STARBUCKS", "MCDONALD'S", "BURGER KING", "TACO BELL", "SUBWAY"])]

nyc_df.head()
```

```python
rest_list_old_score = nyc_df
rest_list_old_score['NY 100 pt score'] = 100-nyc_df['SCORE']
rest_list_old_score.head()
```

```python
NYC_data = rest_list_old_score.drop(['SCORE'], axis=1)
NYC_data.index.names = ['ID']
NYC_data.rename(columns={'NY 100 pt score':'score'}, inplace=True)
NYC_data.head()
```

```python
NYC_Score_Avg = NYC_data.groupby("rest_no")["score"].mean()
NYC_Score_Count = NYC_data.groupby("rest_no")["score"].count()
NYC_Score_Avg
```

```python
nyc_score_df = pd.DataFrame(NYC_Score_Avg)
nyc_score_df.rename(columns={'score':'NYC_score_avg'}, inplace=True)
nyc_score_df
```

```python
nyc_count_df = pd.DataFrame(NYC_Score_Count)
nyc_count_df.rename(columns={'score':'NYC_count'}, inplace=True)
nyc_count_df
```

```python
#austin_count = austin_restaurants['Common Name'].value_counts()
austin_count = Austin_data.groupby("rest_no")["score"].count()
austin_count_df = pd.DataFrame(austin_count)
austin_count_df.rename(columns = {"score": "Austin Count"}, inplace = True)
austin_count_df
```

```python
austin_Score_Avg = Austin_data.groupby("rest_no")["score"].mean()
Austin_score_df = pd.DataFrame(austin_Score_Avg)
Austin_score_df.rename(columns={"score":"Austin_score_avg"}, inplace = True)
Austin_score_df
```

```python
#la_count = la_restaurants['Common Name'].value_counts()
```

```python
la_count = LA_data.groupby("rest_no")["score"].count()
la_count_df = pd.DataFrame(la_count)
la_count_df.rename(columns = {"score": "LA Count"}, inplace = True)
la_count_df
```

```python
la_Score_Avg = LA_data.groupby("rest_no")["score"].mean()
la_score_df = pd.DataFrame(la_Score_Avg)
la_score_df.rename(columns = {"score": "LA_score_avg"}, inplace = True)
la_score_df
```

```python
Austin_avg = pd.merge(Austin_score_df, austin_count_df, on="rest_no", how="left")
Austin_avg
```

```python
LA_avg = pd.merge(la_score_df, la_count_df, on="rest_no", how="left")
LA_avg
```

```python
NYC_avg = pd.merge(nyc_score_df, nyc_count_df, on="rest_no", how="left")
NYC_avg
```

```python
Austin_LA_Merge = pd.merge(LA_avg, Austin_avg, on="rest_no", how="left")
Austin_LA_Merge
```

```python
top_rest = pd.merge(Austin_LA_Merge, NYC_avg, on = "rest_no", how = 'left')
top_rest
```

```python
top_rest['Total_count']=top_rest['LA Count'] + top_rest['Austin Count'] + top_rest['NYC_count']
top_rest = top_rest.drop(['LA Count', 'Austin Count', 'NYC_count'], axis = 1)
top_rest
```
